```python
import pandas as pd
import glob

# Initialize an empty DataFrame
df = pd.DataFrame()

# Get a list of all .dat files in the directory
files = glob.glob('data\hka-aqm-a\*.dat')

# Loop through the list of files
for file in files:
    # Read each file into a DataFrame
    df_temp = pd.read_csv(file, sep=';', skiprows=1, skipinitialspace=True)
    
    # Append the data from the current file to the main DataFrame
    df = pd.concat([df, df_temp], ignore_index=True)
```
