```python
from copy import deepcopy
def build_lvl_df(df, device_ids: list, output_cols: list, reset_ind: bool=True) -> pd.DataFrame:
    """
    args:   df: pd.DataFrame
            device_ids: list
            output_cols: list
            reset_ind: bool

    returns: pd.DataFrame
    """
    df=deepcopy(df)
    return df[df["device_id"].isin(device_ids)][["date_time"]+output_cols].groupby("date_time").agg("mean").reset_index(drop=True) if reset_ind else df[df["device_id"].isin(device_ids)][["date_time"]+output_cols].groupby("date_time").agg("mean")
```

```python
def train_test_split(df, train_test_split_perc: float=0.9, output_columns: list=[]):
    
    split_index = int(len(df) * train_test_split_perc)

    X_train = df[output_columns][:split_index].reset_index(drop=True)
    y_train = df[output_columns][:split_index].reset_index(drop=True)

    X_test = df[output_columns][split_index:].reset_index(drop=True)
    y_test = df[output_columns][split_index:].reset_index(drop=True)

    return X_train, y_train, X_test, y_test
```

```python

def build_lvl_df1(df, device_ids: list, output_cols: list, reset_index: bool=True) -> pd.DataFrame:
    """
    args:   df: pd.DataFrame
            device_ids: list 
            output_cols: list
            reset_index: bool

    returns: pd.DataFrame    
    """
    df = deepcopy(df)
    df_conc=pd.concat([df for df in [df[df["device_id"] == device_id][output_cols].reset_index(drop=True).set_index(["date_time"]) for device_id in device_ids]], axis=1, join="outer")

    for col in output_cols[1:]:
        df_conc[f"{col}_1"] = df_conc.filter(like=col).mean(axis=1)
    
    df_conc=df_conc[[f"{col}_1" for col in output_cols[1:]]].reset_index(drop=True) if reset_index else df_conc[[f"{col}_1" for col in output_cols[1:]]]
    df_conc.columns = df_conc.columns.str.rstrip("_1")
    
    return df_conc
```
